**GET VIDEO AND RUNNING PRODUCER**

```python
#youtube api key
API_KEY = 'AIzaSyD1iqWqp99HOmb2X0V2HdG-sNZ4zcqV9Hs'

def get_recent_videos():
    youtube = build('youtube', 'v3', developerKey=API_KEY)

    while True:
        video_data = {
            'Timestamp': [],
            'Video ID': [],
            'Title': [],
            'Category': [],
            'Duration': [],
            'View Count': []
        }
        # Lấy danh sách 1000 video mới nhất từ khu vực Việt Nam
        request = youtube.search().list(
            part='snippet',
            regionCode='VN',  # Chọn khu vực Việt Nam
            maxResults=50,
            order='date'
        )
        response = request.execute()

        # Lặp qua từng video và trích xuất thông tin
        for item in response['items']:
            video_id = item['id']['videoId']
            video_title = item['snippet']['title']

            # Lấy thông tin chi tiết của video
            video_request = youtube.videos().list(
                part='snippet,statistics,contentDetails',
                id=video_id
            )
            video_response = video_request.execute()

            # Trích xuất thông tin về thể loại, độ dài và số lượt xem của video
            category = video_response['items'][0]['snippet']['categoryId']
            duration = video_response['items'][0]['contentDetails']['duration']
            view_count = video_response['items'][0]['statistics']['viewCount']

            # Lấy timestamp hiện tại
            timestamp = str(datetime.now())

            # Thêm thông tin vào video_data
            video_data['Timestamp'].append(timestamp)
            video_data['Video ID'].append(video_id)
            video_data['Title'].append(video_title)
            video_data['Category'].append(category)
            video_data['Duration'].append(duration)
            video_data['View Count'].append(view_count)

        # Tạo DataFrame từ video_data
        df = pd.DataFrame(video_data)
        # display(df)  # In ra DataFrame để kiểm tra
        # Bạn có thể xử lý DataFrame ở đây, ví dụ: lưu vào file CSV hoặc làm bất kỳ thao tác nào khác cần thiết

        for i in range(50):
            dict_video = df.to_dict(orient="records")[i]
            producer.send('demo_test', value=dict_video)
            sleep(1)

if __name__ == '__main__':
    get_recent_videos()
```
